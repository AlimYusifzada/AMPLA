```python
import ampla
proj=ampla.Proj("D:\\proj\\proj\\proj\\SANGACH3\\NODES\\NODE95\\SRCE") 
for pc in proj.SRCE.keys():
    pcblk=proj.SRCE[pc].Blocks
    for kb in pcblk.keys():
        if pcblk[kb].Name=="VALVECON": # search for valve
            tagname=pcblk[kb].GetPin(":PARAM")[:-8] # get tagname
            if pcblk[kb].GetPin(":T2")[:2]=="D=":
                print(tagname,',\t',pcblk[kb].GetPin(":T2")) # get timer set
            elif pcblk[kb].GetPin(":T2")[:2]=="PC": # it set at the other place (partial stroke option)
                timeadr=pcblk[kb].GetPin(":T2")[:-3]
                print(tagname,',\t',pcblk[timeadr].GetPin(':12'))
```
